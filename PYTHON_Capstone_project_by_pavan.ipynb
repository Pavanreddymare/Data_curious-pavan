```python
## TASK- 2

# loading back saved csv file into dataframe to perform further operations 
import pandas as pd
import numpy as np
Project_df = pd.read_csv('E:\PAVAN\Data_science\PYTHON\Project_df.csv',index_col=0)
Project_df

```

```python
##run average and insert resulting column to dataFrame

run_avg = pd.Series(Project_df['Cost'].rolling(window=3, center=True,min_periods=1).mean())
run_avg = run_avg.astype(float).round()


Project_df.insert(3,"RunAvgCost", run_avg)
Project_df
```

```python
## running For loop to replace the missing values by running average in df

for i, cost in enumerate(Project_df['Cost']):
    if pd.isna(cost):
        Project_df.at[i, 'Cost'] = run_avg[i]
Project_df
```

```python
## TASK-3 :- Split the name column in the Employee dataframe into two new columns “First Name”,
# and “LastName” and remove the older “name” column.
#Importing 'Employees_df'.CSV file

Employee_df = pd.read_csv('E:\PAVAN\Data_science\employee\employee_df.csv',index_col=0)
Employee_df
```

```python
## SPLIT NAMES USING X.SPLIT FUNCTION
names = list(Employee_df['Name'])

first_names = list(map(lambda x : x.split()[0], names))
Last_names = list(map(lambda x: x.split()[1], names))

# INSERT NEWLY CRAETED TABLES TO EMPLOYEE_DF DATAFRAME
Employee_df.insert(1, "First Name", first_names)
Employee_df.insert(2, "Last Name", Last_names)

## DROP EXISTING NAME COLUMN 

Employee_df.drop('Name', inplace = True, axis=1)
```

```python
#FINAL DATAFRAME AFTER WE SPLIT NAMES AND DROP NAME COLUMN 

Employee_df
```

```python
##TASK-4 :-Join all three dataframes in one single dataframe. Name it “Final”


# calling Seniority_df.csv file for remaining dataframe
Seniority_df = pd.read_csv('E:\PAVAN\Data_science\seniority\seniority_level.csv', index_col=0)
Seniority_df
```

```python
# join first two dataframes using sql like join
join_2_df = pd.merge(Employee_df,Seniority_df,on="ID")
join_2_df
```

```python
## joining final dataframe by merging first two and project_df

Final = pd.merge(join_2_df,Project_df,on="ID")
Final
```

```python
## TASK-5 :- Add a new bonus column in the Final dataframe. Give a 5% bonus concerning project
# cost only to employees who have finished the projects.

results_list  = []
for i, r in Final.iterrows():
    if(Final['Status'].iat[i] == 'Finished'):
        result=(r['Cost']) + (r['Cost'])*0.05
        results_list.append(result)
    else:
        result=(r['Cost']) + (r['Cost'])*0
        results_list.append(result)
## Finally insert 'Bonus' column with desired loop result
Final.insert(9, "Bonus", results_list)
```

```python
# Final resultant dataframe with 'bunus' column
Final
```

```python
## TASK - 6 :- Demote the designation level by 1, whose projects have status “fail”. Delete the
# employees record whose designation level is above 4.
new_des = []
for i, r in Final.iterrows():
    if(Final["Status"].iat[i] == 'Failed'):
        result=(r['Designation Level']) - 1
        new_des.append(result)
    else:
        result=(r['Designation Level'])-0
        new_des.append(result)
Final.insert(6,'new_designation',new_des)
```

```python
Final.drop('Designation Level',inplace=True,axis=1)
```

```python
# Final resultant dataframe with 'new_designation' column
Final
```

```python
#DataFrame with updated Designation Levels

Final.rename(columns={'new_designation': 'Designation Level'}, inplace=True)
Final
```

```python
#TASK 7 :- Add “Mr.” and “Mrs.” to the first name column and drop the gender column.
##  here's how we can add 'mr', and 'mrs' to first name column by using for loop
for i, r in Final.iterrows():
    if(Final['Gender'].iat[i] == 'M'):
        Final['First Name'].iat[i] =  'Mr.' + Final['First Name'].iat[i]
    else:
        Final['First Name'].iat[i] =  'Mrs.' + Final['First Name'].iat[i]
```

```python
Final.drop('Gender',inplace=True,axis=1)
```

```python
Final
```

```python
## TASK 8 :- Promote designation level by 1 for the employees whose age is more than 29 years using
# IF condition.

new_des = []

for i, r in Final.iterrows():
    if(Final["Age"].iat[i] > 29):
        result=(r['Designation Level']) + 1
        new_des.append(result)
    else:
        result=(r['Designation Level'])+0
        new_des.append(result)

Final.insert(6, 'Designation', new_des)
Final.drop('Designation Level', inplace = True, axis = 1)
Final.rename(columns={"Designation":"Designation Level"}, inplace = True)
Final
```

```python
## TASK 9:- Add the cost of all projects for each Employee and save it in new dataframe
# “TotalProjCost” with three columns ID, First Name, and Total cost
Final
group = Final.groupby(['ID','First Name'])
```
